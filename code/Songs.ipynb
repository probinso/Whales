```python
import aco
import datetime
import IPython

%matplotlib inline
```

```python
loader = aco.ACOio('../raw/')
parms = dict(frame_duration=.001, frame_shift=.0004)
```

```python
datetime.timedelta(seconds=-1
                  ).total_seconds()
```

```python
date = datetime.datetime(day=19, month=2, year=2016, hour=7, minute=50)
src = loader.load(date)
src.Listen()
snip = src[
    datetime.timedelta(minutes=1, seconds=15):
    datetime.timedelta(minutes=1, seconds=25)]
snip.View()
snip.View('logspectrogram', **parms)
snip.Listen()
```

```python
date = datetime.datetime(day=18, month=2, year=2016, hour=7, minute=55)

src = loader.load(date)
snip = src[
    datetime.timedelta(seconds=7):
    datetime.timedelta(seconds=11)]
snip.View('logspectrogram', **parms)
```

```python
snip.View()

snip.Listen()
```

```python
date = datetime.datetime(day=19, month=2, year=2016, hour=7, minute=30)
src = loader.load(date)

snip = src[
    datetime.timedelta(minutes=2, seconds=40):
    datetime.timedelta(minutes=2, seconds=50)]
snip.View()

snip.View('logspectrogram', **parms)
snip.Listen()
```
